```python
#ประกาศ
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#lab01
sin_sig = thinkdsp.SinSignal(freq=7026, amp=26, offset=0)

sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#lab02
#สามเหลี่ยม
#TriangleSignal(f)
#duration=signal.period*n
signal=thinkdsp.TriangleSignal(4000)
duration=signal.period*2500
segment=signal.make_wave(duration, framerate=10000)
segment.plot()

oowave=signal.make_wave(duration, framerate=10000)
oowave.apodize()
oowave.make_audio()

```

```python
spectrum=oowave.make_spectrum()
spectrum.plot()
```

```python
#ประกาศ4เหลี่ยม
signal = thinkdsp.SquareSignal(4000)
duration = signal.period*2500
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)

wave = signal.make_wave(duration=0.5, framerate=10000)
wave.apodize()
wave.make_audio()

```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```
